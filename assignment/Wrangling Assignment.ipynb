**Q1.** Open the "tidy_data.pdf" document in the repo, which is a paper called Tidy Data by Hadley Wickham.

  1. Read the abstract. What is this paper about?

  This paper is about data cleaning and how to prep data for large-scale analysis. It talks about the best framework to ensure that the data is organized correctly, which is where each variable is a column, each observation is a row, and each type of observational unit is a table.

  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

  The "tidy data standard" is designed to simplify the process of data cleaning by ensuring that the process is consistent and organized. Doing so allows data scientists to focus on the main problem instead of worrying about the intricacies of the data organization.

  3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, itâ€™s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general."

  The first sentence means that the structure for the tidy datasets are all similarly organized, and it is easy to draw information from them, but messy datasets are hard to predict and difficult to understand.

  The second sentence means that the observations and variables are easy to point out, but it is hard to draw connections and define them between datasets since they are all so different.
  4. Read Section 2.2. How does Wickham define values, variables, and observations?

  Wickham defines values as either numbers or strings. Numbers would be quantitative values and strings would be qualitative values. Variables then contain all values that measure the same attribute across units. An observation contains all values measured on the same unit across attributes.
  5. How is "Tidy Data" defined in section 2.3?

  Tidy data is defined so that in tidy datasets, each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
  6. Read the intro to Section 3 and Section 3.1. What are the 5 most common problems with messy datasets? Why are the data in Table 4 messy? What is "melting" a dataset?

  The five most common problems with messy datasets are that:
  - Column headers are values, not variable names.
  - Multiple variables are stored in one column.
  - Variables are stored in both rows and columns
  - Multiple types of observational units are stored in the same table.
  - A single observational unit is stored in multiple tables.

The data in Table 4 is messy because the column headers represent the values of the income and not the variables. This issue can be solved by melting the table, which means turning the columns into rows.

  8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?

  The chicken and the egg problem applies because the tidy data is only as useful as the tools that come with it, which means that if the tools are not used, then there will be no tidy datasets, which would vause data cleaning to become more difficult.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1.
df = pd.read_csv('./data/airbnb_hw.csv')
#print(df.shape, '\n')
df.head(10) #prints the first ten headers
```

```python
price_df = df['Price'] #gets the values of the price column
price_df.unique() #gets the unique values of the price column
```

```python
price_df = price_df.astype(str) # convert the series to a string
price_df = price_df.str.replace(',', '') #removes the commas from the price column
print(price_df.unique()) #prints the unique values of the price column
price_df = pd.to_numeric(price_df, errors='coerce') #converts the price column to a numeric column
print(price_df.unique()) #prints the unique values of the price column
print('total missing: ', sum(price_df.isnull())) #prints the total missing values
```

```python
#2.
df = pd.read_csv('./data/sharks.csv', low_memory=False)
df.head(10)
```

```python
df['Type'].unique()
df['Type'].value_counts()
```

```python
type = df['Type']
type = type.replace(['Sea Disaster', 'Boat', 'Boating', 'Boatomg'],'Watercraft') #replaces the values in the type column
type.unique()
type.value_counts()

type.replace(['Invalid', 'Questionable', 'Unconfirmed', 'Unverified','Under investigation'], np.nan, inplace=True) #replaces the values in the type column
type.unique()
type.value_counts()
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F', 'M', '2017'], np.nan) #replaces the values in the fatal column
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y', 'Y') #replaces the value in the fatal column
pd.crosstab(df['Type'], df['Fatal (Y/N)'], normalize='index') #prints the cross tab of the type and fatal columns
```

```python
#3.
df = pd.read_csv('./data/pretrial_data.csv', low_memory=False)
df.head(10)
```

```python
release = df['WhetherDefendantReceivedPretrialServicesAgencySuperv_PTCC'] #gets the values of the release column
release.unique() #gets the unique values of the release column
print(release.unique(),'\n')
print(release.value_counts(),'\n') #prints the value counts of the release column
release = release.replace(9,np.nan) #replaces the value 9 with nan
print(release.value_counts(),'\n') #prints the value counts of the release column
print(release.unique()) #prints the unique values of the release column
sum(release.isnull()) # 31 missing values
df['WhetherDefendantReceivedPretrialServicesAgencySuperv_PTCC'] = release
del release
```

```python
#4.
length = df['ImposedSentenceAllChargeInContactEvent'] #gets the values of the length column
length.unique() #gets the unique values of the length column
type = df['SentenceTypeAllChargesAtConvictionInContactEvent'] #gets the values of the type column
type.unique() #gets the unique values of the type column

length = pd.to_numeric(length,errors='coerce')
length_NA = length.isnull()
print( np.sum(length_NA),'\n') # many missing

print( pd.crosstab(length_NA, type), '\n')
length = length.mask( type == 4, 0)
length = length.mask( type == 9, np.nan)

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # 274 missing

df['ImposedSentenceAllChargeInContactEvent'] = length
del length, type
```
